## Pandas Dataframe Tool

```python
import pandas as pd
```

```python
data = [1,2,3,4,5,6,7,8,9]
df = pd.DataFrame(data, columns=["x"])
df
```

How to add new columns ?

```python
df["x_plus_5"] = df["x"]+5
df
```

```python
df["x_square"] = df["x"]* df["x"]
df
```

```python
df["is_even"] = df["x_square"] % 2
df
```

### Using Map() function

```python
df["odd_even"] = df["is_even"].map({1:"odd", 0:"even"})
df
```

### Selecting useful columns only

```python
df[["x","x_square"]]
```

### Filtering the features 

```python
df[df["x_square"]>=10]
```

```python
df[(df["x_square"]>=10)|(df["odd_even"]=='odd') ]
```

```python
df[(df["x_square"]>=10)&(df["odd_even"]=='odd') ]
```
